```python
# use FOR loop with Dict keys

for k in myCat.keys():
    print(k)
```

```python
# use FOR loop with Dict values

for k in myCat.values():
    print(k)
```

```python
# use FOR loop to print both KEYS and VALUES

for k, v in myCat.items():
    print(k,v)
```

```python
# Print tuples 

for i in myCat.items():
    print(i)
```

### A tuple is a list which uses parenthesis instead of square brackets.

```python
# Print tuples in LIST. 

newList = []

for i in myCat.items():
    newList.append(i)
```

```python
newList
```

```python
# use GET. Return a substitute value is key doesn't exist 

myCat.get('size', 0)
```

```python
myCat.get('hands', '')
```

### Character Count Program

```python
message = 'It was a bright cold day in April, and the clocks were striking thirteen.'

# List all the letters as keys and count how many of each are in the string
count = {} 

for character in message:
    
    # every character will start out with a 0 count
    count.setdefault(character, 0)
    
    # update the count of characters as the program loop through the string
    count[character] = count[character] + 1
    
print(count)
```

### Change message to uppercase

```python
message = 'It was a bright cold day in April, and the clocks were striking thirteen.'

# List all the letters as keys and count how many of each are in the string
count = {} 

for character in message.upper():
    
    # every character will start out with a 0 count
    count.setdefault(character, 0)
    
    # update the count of characters as the program loop through the string
    count[character] = count[character] + 1
    
print(count)
```

### Import module to clean up print out

```python
import pprint

message = 'It was a bright cold day in April, and the clocks were striking thirteen.'

# List all the letters as keys and count how many of each are in the string
count = {} 

for character in message.upper():
    
    # every character will start out with a 0 count
    count.setdefault(character, 0)
    
    # update the count of characters as the program loop through the string
    count[character] = count[character] + 1
    
pprint.pprint(count)
```

```python
# Print out as a STRING

pprint.pformat(count)
```
