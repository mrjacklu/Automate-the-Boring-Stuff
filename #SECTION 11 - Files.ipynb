```python
# Read File

helloFile = open('hello.txt')
content = helloFile.read()
```

```python
print(content)
```

```python
# Close File

helloFile.close()
```

```python
# Read Lines in a List

helloFile = open('hello.txt')
helloFile.readlines()
```

```python
# Write Mode. Destroys all text and over-writes file.

helloFile = open('hello.txt', 'w')
```

```python
# Append Mode. Continues whatever is inside.

helloFile = open('hello.txt', 'a')
```

# Shelve Module

Used to store complex data structures. Save Python to Binary Shelf Files. Store it, then open then in the future.

Shelf Files are similar to Keys:Value 

```python
import shelve
```

```python
shelfFile = shelve.open('mydata')
shelfFile['cats'] = ['Sophie', 'Pooka', 'Simon']
shelfFile.close()
```

```python
shelfFile = shelve.open('mydata')
shelfFile['cats']
```

```python
shelfFile = shelve.open('mydata')
shelfFile.keys()
```

```python
# List the Keys

list(shelfFile.keys())
```

```python
# List the Values

list(shelfFile.values())
```
