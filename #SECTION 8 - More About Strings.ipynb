### Join Method combines string

```python
', '.join(['cats','rats','bats'])
```

```python
'\n\n'.join(['cats','rats','bats'])
```

```python
print('\n\n'.join(['cats','rats','bats']))
```

### Split Method

```python
# Splits string on white space

'My name is Simon'.split()
```

```python
# Split on other characters

'My name is Simon'.split('m')
```

### LJust( ) / Rjust( )  / Center( ) - adds padding

```python
'Hello'.rjust(10)
```

```python
len('Hello'.rjust(10))
```

```python
'Hello'.ljust(10)
```

```python
'Hello'.rjust(10,'-')
```

```python
'Hello'.center(10)
```

```python
'Hello'.center(20, '=')
```

### Lstrip / Rstrip

```python
'    Hello'.strip()
```

```python
# Remove spacing from leftside of the string

'    Hello   '.lstrip()
```

```python
'SpamSpamSpamBaconEggsEggsSpamSpam'.strip('ampS')
```
