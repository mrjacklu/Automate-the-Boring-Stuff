### Keyword Arguments - optional

```python
# The print function by default adds a new line character. 

print('hello')
print('world')
```

```python
# But you can change this by adding in the optional argument

print('hello', end=' ')
print('world')
```

```python
# But you can also include multiple strings in Print

print('apple', 'banana', 'orange')
```

```python
# There's more! Let's include the sep keyword.

print('apple', 'banana', 'orange', sep = ' > ')
```

## 10. Global and Local Scopes

Variables in GLOBAL Scope is created and destroyed after program ends.


Variables in local scope are temporary and destroyed after function is called.

• Code in a local scope can access global variable.

• Code in global scope CANNOT acces variable in local scope.

• Code in one function's local cannot use variables in another local scope

• Use same variable names if defined in different local scopes

• Scope can be thought of as an area of the source code, and as a container of variables

```python
# Global Variable

spam = 42
```

```python
# Local Variable

def egg():
    spam = 42
```

```python
# Using same variable name in local scopes

def spam():
    eggs = 99
    bacon()
    print(eggs)

def bacon():
    ham = 101
    eggs = 80
    print(ham)
```

```python
spam()
```
